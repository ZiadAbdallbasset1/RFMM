```python
df.dropna(subset=['CustomerID'], inplace=True)
```

```python
df
```

```python
df = df[~df['InvoiceNo'].str.contains('C', na=False)]  # Remove cancelled transactions
```

```python
df
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df['reveneue'] = df['UnitPrice'] * df['Quantity']
```

```python
max_date = df['InvoiceDate'].max()
```

```python
rfm = df.groupby('CustomerID').agg({
   'InvoiceDate': 'max',
         'InvoiceNo': 'count',
         'reveneue': 'sum'
      })
```

```python
rfm.rename(columns={
  'InvoiceDate': 'last_purchase_date',
      'InvoiceNo': 'frequency',
      'reveneue': 'monetary_value'
}, inplace=True)
```

```python
rfm
```

```python
rfm['recency'] = (max_date - rfm['last_purchase_date']).dt.days
```

```python
rfm.drop(columns=['last_purchase_date'], inplace=True)
```

```python
rfm = rfm[['recency', 'frequency', 'monetary_value']]
```

```python
rfm['R'] = pd.qcut(rfm['recency'], 4, labels=[4, 3, 2, 1])
rfm['F'] = pd.qcut(rfm['frequency'], 4, labels=[1, 2, 3, 4])
rfm['M'] = pd.qcut(rfm['monetary_value'], 4, labels=[1, 2, 3, 4])
rfm['RFM_Score'] = rfm['R'].astype(str) + rfm['F'].astype(str) + rfm['M'].astype(str)
```

```python
rfm
```

```python
rfm['T-score'] = rfm[['R','F','M']].sum(axis=1)
```

```python

   
def segment(row):
     if row['T-score'] <= 5:
          return 'low value'
     elif row['T-score'] <= 9:
          return 'middle value'
     else:
          return 'high value'
rfm['segment'] = rfm.apply(segment, axis=1)
```

```python
rfm
```

```python
fig = px.bar(rfm, x='segment', y='T-score', color='segment',
      title='Customer Segmentation by RFM Score',
      labels={'T-score': 'Total RFM Score', 'segment': 'Customer Segment'},
      color_continuous_scale=colors.qualitative.Plotly)
fig.update_traces(marker_opacity=1)
fig.update_layout(template="plotly_white")
fig.show()
```
